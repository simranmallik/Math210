### Q3

Let's see how well an edge detection function can reveal the features of your picture.  
1. Copy the `edges()` function from the prelab to a code block.
2. Apply `edges()` to your image and save the result to a variable.


```
def edges(xy):
  dfdy = np.roll(xy,1,axis=0) - xy
  dfdx = np.roll(xy,1,axis=1) - xy
  new_array = (dfdy**2 + dfdx**2)**(1/2)
  return new_array
  print(new_array)

edges_mona = edges(mona)
```
