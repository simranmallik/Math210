## Q1 
* Copy/paste your `equilibrium(f,a,n)` function from the prelab to a code cell and run it.  
* Use it to graph the first `100` terms of the recursive sequence defined by your function, with initial value `0.5271`.  This value is unlikely to be at an equilibrium by coincidence.  
* What equilibrium does your series appear to approach?  Check with your group and see that all of your functions exhibit the same behavior for `k=0.5`



```
def equilibrium(f,a,n):
  sequence = [a] 
  for i in range(n):
    a = f(a)
    sequence.append(a)
  x = np.arange(0,1,.01)
  y = f(x)
  fig, axs = plt.subplots(1,2, figsize = (10,5))
  axs[0].plot(x,y)
  y = x
  axs[0].set_aspect('equal')
  axs[0].plot(x,y)
  axs[0].grid()
  axs[1].plot(sequence, 'o')
  axs[1].grid()
  #axs[1].set_ylim([-.05,.05])
  plt.show()
  return sequence[n]

equilibrium(f2, .5271, 100)
```

The series approaches 0.0.
